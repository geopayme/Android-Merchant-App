## 1. Ferrofluid Vortex Formation
**Objective:** Investigate the formation of stable vortices in ferrofluids under rotation and magnetic fields.

**Instructions:** Input your data for rotation speed and observed vortex count.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Input your data here
vortex_data = pd.DataFrame({
    'Rotation Speed (RPM)': [2, 4, 6, 8, 10],
    'Vortex Count': [0, 2, 3, 5, 7]
})
vortex_data
```

```python
# Plotting vortex formation
plt.plot(vortex_data['Rotation Speed (RPM)'], vortex_data['Vortex Count'], marker='o')
plt.title('Vortex Formation vs Rotation Speed')
plt.xlabel('Rotation Speed (RPM)')
plt.ylabel('Vortex Count')
plt.grid(True)
plt.show()
```

## 2. Ferrofluid Pattern Formation
**Objective:** Analyze pattern emergence in response to vertical magnetic fields.

**Instructions:** Record and analyze fractal dimension vs magnetic field strength.

```python
# Input your data for magnetic field strength and fractal dimension
pattern_data = pd.DataFrame({
    'Magnetic Field (mT)': [100, 200, 300, 400, 500],
    'Fractal Dimension': [1.2, 1.4, 1.6, 1.7, 1.8]
})
pattern_data
```

```python
# Plotting fractal dimension vs magnetic field strength
plt.plot(pattern_data['Magnetic Field (mT)'], pattern_data['Fractal Dimension'], marker='s', color='green')
plt.title('Fractal Dimension vs Magnetic Field')
plt.xlabel('Magnetic Field (mT)')
plt.ylabel('Fractal Dimension')
plt.grid(True)
plt.show()
```

## 3. Droplet Manipulation
**Objective:** Study ferrofluid droplet response to controlled magnetic fields.

**Instructions:** Input droplet size and resulting displacement under magnetic control.

```python
# Input droplet data
droplet_data = pd.DataFrame({
    'Droplet Volume (uL)': [1, 2, 3, 4, 5],
    'Displacement (mm)': [0.5, 1.2, 2.0, 2.8, 3.5]
})
droplet_data
```

```python
# Plotting droplet displacement
plt.plot(droplet_data['Droplet Volume (uL)'], droplet_data['Displacement (mm)'], marker='^', color='red')
plt.title('Droplet Displacement vs Volume')
plt.xlabel('Droplet Volume (uL)')
plt.ylabel('Displacement (mm)')
plt.grid(True)
plt.show()
```

## 4. Turbulence Analysis
**Objective:** Explore flow instability and turbulence under magnetic agitation.

**Instructions:** Record Reynolds number and qualitative turbulence level.

```python
# Turbulence data
turbulence_data = pd.DataFrame({
    'Reynolds Number': [100, 300, 500, 700, 900],
    'Turbulence Intensity (0-1)': [0.1, 0.25, 0.5, 0.75, 0.95]
})
turbulence_data
```

```python
# Turbulence plot
plt.plot(turbulence_data['Reynolds Number'], turbulence_data['Turbulence Intensity (0-1)'], marker='D', color='purple')
plt.title('Turbulence Intensity vs Reynolds Number')
plt.xlabel('Reynolds Number')
plt.ylabel('Turbulence Intensity')
plt.grid(True)
plt.show()
```
